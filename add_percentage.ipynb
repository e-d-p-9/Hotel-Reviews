```python
import pandas as pd
```

```python
# read in the hotel sub1 with class labels and character numbers from numbers on mac
sub1_class_chars = pd.read_csv('/Users/ethan/GitHub_Desktop/Hotel-Reviews/task2/hotelsub1_index_sort_class.csv')
# create a new dataframe for appending to
new_df = pd.DataFrame(columns=['class','part', 'Annotations', 'Num Characters', 'Percentage'])
# get all unique class labels as a list
class_labels = sub1_class_chars['class'].unique()

# add percent labels to all
for i in range(len(class_labels)):
    # grab current class
    current_class = sub1_class_chars.loc[sub1_class_chars['class'] == class_labels[i]]

    # add percentage column unpopulated to current class selection
    current_class['Percentage'] = ''

    # sort based on num characters
    current_class = current_class.sort_values(by="Num Characters")

    # insert percentage labels
    percent = 20
    for i in range(5):
        current_class.iloc[i, current_class.columns.get_loc('Percentage')] = percent
        percent += 20


    # add to new dataframe
    new_df = pd.concat([new_df, current_class])

new_df.to_csv('test1.csv')
```

```python
#list to hold percentages parallel to sub1 iteration
percent_list = []
for i in range(len(testsub)):
    # get current rows index
    current_index= testsub['index'].iloc[i]
    # get text count of part
    char_count = len(testsub['part'].iloc[i])
    # find row in reviews
    reviews_row = reviews[reviews['Index'] == current_index]
    # get character count of reviews text
    rev_char_count = len(reviews_row['reviews.text'].iloc[0])
    
    # divide to get percentage
    percentage = char_count / rev_char_count
    # add to list
    percent_list.append(percentage)
    
    print(f"character count: {char_count}")
    print(rev_char_count)
    print('------------------')
    


print(percent_list)

# add percentage column to sub1
testsub.insert(3, "Percent", percent_list, True)

testsub.to_csv("testpercentage.csv")
```
